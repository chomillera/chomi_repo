```python
# import library

import pandas as pd
import numpy as np
import plotly as plt
import plotly.express as px
from tabulate import tabulate as tbl
```

```python
data = pd.read_csv('covid_19_indonesia_time_series_all.csv')
```

# Data Understanding & Cleaning

```python
data.info()
```

```python
# Drop column yang tidak diperlukan

data.drop(['City or Regency', 'Special Status'], axis=1, inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.info()
```

```python
df = data[['Date', 'Location', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Location Level', 'Province', 'Country', 'Population']]
```

```python
df = df.query("Location != 'Indonesia'")
```

```python
df.info()
```

```python
df.dtypes
```

```python
# Merubah kolom date ke datetime
df.loc[:, 'Date'] = pd.to_datetime(df['Date'])
```

```python
fig = px.line(df, x='Date', y='New Cases', title='Perkembangan New Cases Berdasarkan Tanggal')
fig.show()
```

```python
# Menghitung total kematian di setiap provinsi
total_death_per_province = df.groupby('Province')['New Deaths'].sum().reset_index()

# Mengurutkan data berdasarkan total kematian dari yang terbesar
total_death_per_province_sorted = total_death_per_province.sort_values(by='New Deaths', ascending=False)

# Membuat grafik bar menggunakan Plotly Express (px)
fig = px.bar(total_death_per_province_sorted, x='Province', y='New Deaths', title='Total Kematian per Provinsi')
fig.show()
```

```python
# Karena data yang ditampilkan kurang enak dipandang, maka akan dilakukan beberapa adjustment
# Memilih top 10 provinsi saja
total_death_per_province_top = total_death_per_province_sorted.head(10)

# Melakukan beberapa adjustment pada grafik agar datanya lebih enak dilihat
fig = px.bar(total_death_per_province_top, 
             x='Province',
             y='New Deaths',
             title='10 Provinsi dengan Jumlah Kematian Terbanyak',
             labels={'New Deaths':'Jumlah Kematian','Province':'Provinsi'},
             color='New Deaths',
             text='New Deaths',
             color_continuous_scale=['yellow','red'])

fig.update_layout(
    coloraxis_showscale=False, # Menonaktifkan skala warna jika tidak diperlukan
    plot_bgcolor='white' # Background putih
)
```

```python
# Ratio antara kematian dengan kesembuhan

# Mengelompokkan data berdasarkan 'Province' dan menghitung total 'New Deaths' dan 'New Recovered'
grouped_data = df.groupby('Province').agg({'New Deaths': 'sum', 'New Recovered': 'sum'}).reset_index()

# Menghindari pembagian dengan nol
grouped_data = grouped_data[grouped_data['New Recovered'] != 0]

# Menghitung rasio total kematian terhadap total kesembuhan untuk setiap provinsi
grouped_data['Death to Recovery Ratio'] = grouped_data['New Deaths'] / grouped_data['New Recovered']

# Menampilkan hasil perhitungan
grouped_data.head()
```

```python
total_new_cases_per_province = df.groupby('Province')['New Cases'].sum().reset_index()
total_new_cases_per_province_sorted = total_new_cases_per_province.sort_values('New Cases', ascending=False)
total_new_cases_per_province_top = total_new_cases_per_province_sorted.head(10)

fig = px.bar(total_new_cases_per_province_top,
             x='Province',
             y='New Cases',
             title='Kasus Covid-19 baru per Provinsi',
             labels={'New Cases':'Kasus Baru','Province':'Provinsi'},
             color='New Cases',
             color_continuous_scale=['yellow','red'],
             text='New Cases')

fig.update_layout(
    coloraxis_showscale=False
)

fig.show()
```
